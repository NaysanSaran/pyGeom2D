```python
import math
import numpy as np

from pygeom import Axes, Point, Vector
```

```python
# Create the cartesian axis
axes = Axes(xlim=(-2,5), ylim=(-3,5), figsize=(10,8))

# vector addition
v0 = Vector(Point(0, 0), Point(-1, 2), label='v0')
v1 = Vector(Point(1, 1), Point(3, 1), label='v1', color='#0e2f44')
v2 = Vector(Point(3, 1), Point(3, 4), label='v2', color='#f6546a')
v3 = v1 + v2
axes.addMany([v0, v1, v2, v3])

# vector subtraction
v4 = v1 - v2
axes.add(v4, linewidth=3, linestyle='dashed')

# vector scalar multiplication
v5 = -0.5*v1
axes.add(v5, color='red')

# get unit vector associated with v1 + v2
v6 = v3.get_unit_vector()
axes.add(v6, color='green', linewidth=2)

# get the additive inverse of a vector
p  = Point(4,1, label='p', color='red')
v7 = Vector(Point(4, 1), Point(4.5, 2), label='v7', linewidth=2)
v8 = -v7

axes.addMany([p, v7, v8])

axes.draw()
```
