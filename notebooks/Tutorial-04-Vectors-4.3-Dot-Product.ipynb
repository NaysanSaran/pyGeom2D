```python
import math
import numpy as np

from pygeom import Axes, Point, Vector
```

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,5), ylim=(-3,5), figsize=(10,8))

# create v1
v1 = Vector(Point(1, 1), Point(4, 1), label='v1', color='#0e2f44', linewidth=2)
# create v2 perpendicular to v1
v2 = Vector(Point(4, 1), Point(4, 3), label='v2', color='#f6546a', linewidth=2) 
# create v3 parallel to v1
v3 = Vector(Point(1, 1), Point(-0.5, 1), label='v3', color='#60a143', linewidth=2) 
v4 = v1 + v2
axes.addMany([v1, v2, v3, v4])

# angle between v1 and v2
print("Angle between v1 and v2 is %.2f degrees" % v1.get_angle(v2))
print("Dot product between v1 and v2 is %.2f"   % v1.dot(v2))
print("Is v1 perpendicular to v2? Answer = %s"  % v1.is_perpendicular(v2))
print("Is v1 parallel to v2? Answer = %s"       % v1.is_parallel(v2))

# angle between v1 and v3
print("\n")
print("Angle between v1 and v3 is %.2f degrees" % v1.get_angle(v3))
print("Dot product between v1 and v3 is %.2f"   % v1.dot(v3))
print("Is v1 perpendicular to v3? Answer = %s"  % v1.is_perpendicular(v3))
print("Is v1 parallel to v3? Answer = %s"       % v1.is_parallel(v3))

axes.draw()
```
