```python
from pygeom import Axes, Point
```

## Basic style (arrows)

```python
# Create the cartesian axis
axes = Axes(xlim=(-5,5), ylim=(-3,5), figsize=(9,7))

point = Point(3, 2, color='#bb4406', label='p', coords=True)

axes.add(point)
axes.draw('./img/tutorial00-basic-style.png')
```

## Custom axes names

```python
# Create the cartesian axis
axes = Axes(xlim=(-5,5), ylim=(-3,5), figsize=(9,7), style=['x', 'f(x)'])

point = Point(3, 2, color='#bb4406', label='p', coords=True)

axes.add(point)
axes.draw('./img/tutorial00-custom-style.png')
```
