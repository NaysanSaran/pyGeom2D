```python
import math
import numpy as np

from pygeom import Axes, Point, Curve
```

## 3.1 Drawing some curves

```python
def f(x):
    y = math.pow(x, 2)
    return y
    
def g(x):
    if x < 0:
        y = np.nan
    else:
        y = math.sqrt(x)
    return y

def h(x):
    y = 1/x
    return y
```

```python
# Create the cartesian axis
axes = Axes(xlim=(-5,5), ylim=(-4,5.2), figsize=(10,9))

# Curve of y = x^2
c1 = Curve(f, label=r'$y=x^2$', color='grey')
axes.add(c1)

# Curve of y = sqrt(x)
c2 = Curve(g, label=r'$y=\sqrt{x}$', linewidth=2, color='#008080')
axes.add(c2)

# Curve of y = 1/x
c3 = Curve(h, x_disc=[0.0], label='y=1/x', linewidth=2, color='#ca2c40')
axes.add(c3)

axes.draw()
```

## 3.2 Approximate tangent lines to curves

```python
# Create the cartesian axis
axes = Axes(xlim=(-5,5), ylim=(-4,5.2), figsize=(10,8))

# Curve of y = x^2 and tangent line at (-1.5,(-1.5)^2)
c1 = Curve(f, label=r'$y=x^2$', color='grey')
p1 = Point(-1.5, f(-1.5), color='red', label='p1')
t1 = c1.approximate_tangent(p1, color='grey', linestyle='dashed', label='tangent at p1', loc='down')
axes.addMany([c1, p1, t1])

# Curve of y = sqrt(x) and tangent line at (2, sqrt(2))
c2 = Curve(g, label=r'$y=\sqrt{x}$', linewidth=2, color='#008080')
p2 = Point(2, g(2), color='red', label='p2')
t2 = c2.approximate_tangent(p2, color='#008080', linestyle='dashed', label='tangent at p2', loc='down')
axes.addMany([c2, p2, t2])

axes.draw()
```

## 3.3 Approximate normal lines to curves

```python
# Create the cartesian axis
axes = Axes(xlim=(-5,5), ylim=(-4,5.2), figsize=(10,8))

# Curve of y = x^2 and tangent line at (-1.5,(-1.5)^2)
c1 = Curve(f, label=r'$y=x^2$', color='grey')
p1 = Point(-1.5, f(-1.5), color='red', label='p1')
t1 = c1.approximate_normal(p1, color='grey', linestyle='dashed', label='normal at p1', loc='down')
axes.addMany([c1, p1, t1])

# Curve of y = sqrt(x) and tangent line at (2, sqrt(2))
c2 = Curve(g, label=r'$y=\sqrt{x}$', linewidth=2, color='#008080')
p2 = Point(2, g(2), color='red', label='p2')
t2 = c2.approximate_normal(p2, color='#008080', linestyle='dashed', label='normal at p2', loc='down')
axes.addMany([c2, p2, t2])

axes.draw()
```
