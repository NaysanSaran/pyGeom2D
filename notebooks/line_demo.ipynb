```python
from pygeom import Axes, Point, Line
```

## Add line, given 2 points

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,7), ylim=(-1,7), figsize=(7,6))

# Points
p1 = Point(1, 1, color='red')
p2 = Point(5, 4, color='green')

l = Line(p1=p1, p2=p2)

axes.addMany([p1, p2, l])
axes.draw()
```

```python
print(l)
```

## Add line, given point & slope

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,7), ylim=(-1,7), figsize=(7,6))

# Points
p = Point(1, 1, color='red')
slope = 0.750 

l = Line(p=p1, slope=slope)

axes.addMany([p, l])
axes.draw()
```

```python
print(l)
```

## Add line, given equation slope & intercept

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,7), ylim=(-1,7), figsize=(7,6))

# Line
l = Line(slope=0.75, intercept=0.25)

axes.add(l)
axes.draw()
```
