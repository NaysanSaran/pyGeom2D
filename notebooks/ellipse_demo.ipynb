```python
import numpy as np
from pygeom import Axes, Point, Ellipse

%matplotlib inline
```

## Demo

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,10), ylim=(-1,10), figsize=(12,10))

# Points
mu = Point(5, 5, color='black', add_coordinates=False)
sigma = np.array([
    [3.0, 1.0],
    [1.0, 2.0]
])

el = Ellipse(mu, sigma)

axes.addMany([mu, el, el.v_major, el.v_minor])
axes.draw()
```
