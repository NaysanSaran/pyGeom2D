```python
import math
import numpy as np

from pygeom import Axes, Point, Vector, Segment
```

### 4.4.1 Vector Projection

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,5), ylim=(-1,5), figsize=(10,8))

# create v1 and v2
v1 = Vector(Point(1, 1), Point(4, 1), label='v1', color='#0e2f44')
v2 = Vector(Point(1, 1), Point(3, 2), label='v2', color='#f6546a') 
axes.addMany([v1, v2])

# projection of v2 over v1
v2_proj = v2.project_onto(v1)
axes.add(v2_proj, label='v2 proj', linestyle='dashed')

s1 = Segment(v2.p2, v2_proj.p2, color='#333333', linestyle='dashed')
axes.add(s1)

# create v3 and v4
v3 = Vector(Point(1, 2), Point(3, 2.5), label='v3', color='#0e2f44')
v4 = Vector(Point(1, 2), Point(2, 5), label='v4', color='#f6546a')
axes.addMany([v3, v4])

# projection of v3 over v4
v3_proj = v3.project_onto(v4)
axes.add(v3_proj, label='v3 proj', linestyle='dashed')

s2 = Segment(v3.p2, v3_proj.p2, color='#333333', linestyle='dashed')
axes.add(s2)

axes.draw()
```

### 4.4.2 Vector parallel to another, of specific norm, starting or ending at specified location 

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,5), ylim=(-1,5), figsize=(10,10))

# create vector and a couple of points
v1 = Vector(Point(1, 1), Point(3, 2), label='v2', color='#f6546a', linewidth=2) 
p1 = Point(1, 2.5, label='p1', loc='down')
p2 = Point(4.5, 0.5, label='p2', color='#09a837')

axes.addMany([v1, p1, p2])

# create vectors parallel to v1
v2 = v1.get_parallel(start=p1, norm=1, label='// to v1, unit norm')
v3 = v1.get_parallel(end=p2, label='// to v1, ending at p2', color='#09a837')
v4 = v1.get_parallel(label='defaut // to v1', color='#181d27', linewidth=2, linestyle='dashed')

axes.addMany([v2, v3, v4])

axes.draw()
```

### 4.4.3 Vector perpendicular to another, starting or ending at specified location 

```python
# Create the cartesian axis
axes = Axes(xlim=(-1,5), ylim=(-1,5), figsize=(10,10))

# create vector and a couple of points
v1 = Vector(Point(1, 1), Point(3, 2), label='v2', color='#f6546a', linewidth=2) 
p1 = v1.get_midpoint(color='blue')
p2 = Point(4, 3.5, label='p2', color='#09a837')

axes.addMany([v1, p1, p2])

# create vectors parallel to v1
v2 = v1.get_perpendicular(start=p1, norm=1, label='v2')
v3 = v1.get_perpendicular(end=p2, label='v3', color='#09a837')
v4 = v1.get_perpendicular(label='v4', color='#181d27', linewidth=2, linestyle='dashed')

axes.addMany([v2, v3, v4])

axes.draw()
```
