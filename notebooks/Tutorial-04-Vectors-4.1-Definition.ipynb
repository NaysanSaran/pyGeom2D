```python
import math
import numpy as np

from pygeom import Axes, Point, Vector
```

```python
# Create the cartesian axis
axes = Axes(xlim=(-5,5), ylim=(-4,5), figsize=(10,8))

# Define a vector using two points
v1 = Vector(Point(1,1), Point(4,2), label='v1', color='grey')
axes.add(v1)

# vertical vector
p3 = Point(-2, 0)
p4 = Point(-2, 4)
v2 = Vector(p3, p4, label='v2', color='#0e2f44')
axes.add(v2)

# another vertical vector
p3 = Point(4, 0)
p4 = Point(4, -1)
v3 = Vector(p3, p4, label='v3', color='#f6546a')
axes.add(v3)

# horizontal vector
p5 = Point(-3, -3, label='p5', color='green')
p6 = Point( 3, -3, label='p6', color='green')
v4 = Vector(p5, p6, label='v3 between p5 and p6', color='green')
axes.addMany([p5, p6, v4])

# another horizontal vector
p7 = Point( 1.5, 1.5, label='p7', color='orange')
p8 = Point(-1.5, 1.5, label='p8', color='orange')
v5 = Vector(p7, p8, label='v4 between p7 and p8', color='orange')
axes.addMany([p7, p8, v5])

axes.draw()
```
